```python
#Using Matplotlib to plot visuals, we analyse Total revenue per month
# Convert 'OrderDate' column to datetime
merged_df['OrderDate'] = pd.to_datetime(merged_df['OrderDate'])

# Create a new column for the 7-month period
merged_df['SevenMonthPeriod'] = merged_df['OrderDate'].dt.to_period('7M')

# Group by 'TwoMonthPeriod' and sum 'TotalRevenue' for each 2-month period
seven_monthly_revenue = merged_df.groupby('SevenMonthPeriod')['Revenue'].sum()

# Convert the 'TwoMonthPeriod' index to strings
seven_monthly_revenue.index = seven_monthly_revenue.index.strftime('%Y-%m')

# Plot the line chart for every 7 months
plt.figure(figsize=(10, 6))
plt.plot(seven_monthly_revenue.index, seven_monthly_revenue.values, marker='o', linestyle='-')
plt.xlabel('7-Month Period')
plt.ylabel('Revenue')
plt.title('Total Revenue Every 7 Months')
plt.xticks(rotation=45)
plt.grid(False)
plt.show()
```

```python
#Using Matplotlib to plot visuals, we analyse Total revenue per month
# Convert 'OrderDate' column to datetime
merged_df['OrderDate'] = pd.to_datetime(merged_df['OrderDate'])

# Create a new column for the 2-month period
merged_df['TwoMonthPeriod'] = merged_df['OrderDate'].dt.to_period('2M')

# Group by 'TwoMonthPeriod' and sum 'OrderQuantity' for each 2-month period
two_monthly_revenue = merged_df.groupby('TwoMonthPeriod')['OrderQuantity'].sum()

# Convert the 'TwoMonthPeriod' index to strings
two_monthly_revenue.index = two_monthly_revenue.index.strftime('%Y-%m')

# Plot the line chart for every 2 months
plt.figure(figsize=(10, 6))
plt.plot(two_monthly_revenue.index, two_monthly_revenue.values, marker='o', linestyle='-')
plt.xlabel('2-Month Period')
plt.ylabel('Order Quantity')
plt.title('Order Quantity Every 2 Months')
plt.xticks(rotation=45)
plt.grid(False)
plt.show()
```

```python
# Calculate churn rate
churn_rate = Sales_all.groupby('TwoMonthPeriod')['CustomerKey'].nunique().pct_change()
churn_rate = churn_rate.fillna(0)  # Replace NaN with 0 for the first period
print(churn_rate)
```

```python

# Calculate customer retention rate
retention_rate = 1 - churn_rate
print(retention_rate)
```

```python

# Calculate customer lifetime value (CLTV)
Sales_all['CLTV'] = Sales_all.groupby('CustomerKey')['Revenue'].transform('sum')

```

```python
# Calculate average CLTV
average_cltv = Sales_all.groupby('TwoMonthPeriod')['CLTV'].mean()
print(average_cltv)
```
